## KNN Demo

```python
from numpy import *
import operator
from os import listdir
```

```python
def creatDataset():
    group = array([[1.0, 1.1],[1.0, 1.0],[0,0],[0, 0.1]])
    labels = ['A', 'A', 'B', 'B']
    return group, labels
```

```python
def classify0(x, dataset, labels, k):
    '''
    Args:
      x: 待分类数据
      dataset: 训练数据
      labels: 标签
      k: k近邻
    '''
    dataSize = dataset.shape[0]
    # 接下来计算两点间距离
    # tile函数为重复某个数组，下面一行代表生成了二维数组
    diffMat = tile(x, (dataSize, 1)) - dataset
    sqDiffMat = diffMat**2
    sqDistance = sqDiffMat.sum(axis=1)
    distance = sqDistance**0.5
    # 对距离进行排序，返回排序后点的下标
    sortedDistIndices = distance.argsort()
    classCount = {}
    for i in range(k):
        # 得到前k个点对应的label
        voteLabel = labels[sortedDistIndices[i]]
        # 统计该类别出现的次数
        # 参数 0 只在初始化时起作用
        classCount[voteLabel] = classCount.get(voteLabel, 0) + 1
    # 对给类别出现的次数进行排序
    sortedClassCount = sorted(classCount.items(),
                             key = operator.itemgetter(1), # 利用字典的第二个的元素进行排序
                             reverse = True)
    # 返回次数最多的label
    return sortedClassCount[0][0]
```

```python
print(classify0([0,0], creatDataset()[0], creatDataset()[1], 3))
```

## Example: Improving matches from a dating site

```python
def readFile(filename):
    f = open(filename)
    dataSet = f.readlines()
    numrow = len(dataSet)
    # 创建值为0的矩阵
    Mat = zeros((numrow, 3))
    labels = []
    index = 0
    for line in dataSet:
        arr = line.strip().split('\t') # 去掉首尾空白字符并分割字符
        Mat[index, :] = arr[0:3] # 将数据保存到矩阵中
        labels.append(int(arr[-1])) # 保存标签
        index += 1
    return Mat, labels
        
```

分别取数据集的一，二列和二，三列进行可视化

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax1 = fig.add_subplot(211)
Mat, labels = readFile('datingTestSet2.txt')
ax1.scatter(Mat[:,1], Mat[:,2], 15.0*array(labels), 15.0*array(labels))
ax1.set_xlabel('Percentage of time spent on video games')
ax1.set_ylabel('liters of ice cream consumed per week')
ax2 = fig.add_subplot(212)
ax2.scatter(Mat[:,0], Mat[:,1], 15.0*array(labels), 15.0*array(labels))
ax2.set_xlabel('Frequent flying miles earned per year')
ax2.set_ylabel('Percentage of time spent on video games')
plt.show()
```

## Example: Digit classification

下面对一个数字数据集进行分类的示例。该数据集分为训练集和测试集，文件以'label_index。txt' 格式保存。txt文件中，数字以32×32的0,1矩阵表示。

```python
def img2vec(filename):
    '''
    将digit转换为一维向量表示
    '''
    digitVec = zeros((1, 1024))
    f = open(filename)
    for i in range(32):
        line = f.readline()
        for j in range(32):
            digitVec[0, 32*i+j] = int(line[j])
    return digitVec
```

```python
def digitTest():
    trainingFiles = listdir('digits/trainingDigits')
    trainSize = len(trainingFiles)
    trainMat = zeros((trainSize, 1024))
    labels = []
    for i in range(trainSize):
        filename = trainingFiles[i].split('.')[0]
        label = int(filename.split('_')[0])
        labels.append(label)
        trainMat[i,:] = img2vec('digits/trainingDigits/%s.txt' % filename)
    
    testingFiles = listdir('digits/testDigits')
    testSize = len(testingFiles)
    errorCount = 0
    for i in range(testSize):
        filename = testingFiles[i].split('.')[0]
        label = int(filename.split('_')[0])
        testVec = img2vec('digits/testDigits/%s.txt' % filename)
        predict = classify0(testVec, trainMat, labels, 3)
        if (predict != label):
            errorCount += 1
    print('total error rate is %f' % (errorCount/float(testSize)))
        
```

```python
digitTest()
```
